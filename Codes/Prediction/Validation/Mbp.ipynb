```python
import pandas as pd
file_path_fimo = "Data/FIMO/result_all_transcripts/fimo.txt"
fimo_df = pd.read_csv(file_path_fimo, sep='\t')
```

```python
print(fimo_df.columns.tolist())
```

```python
# Define the target motif
target_motif = "mmu-miR-466n-5p"

# Filter the DataFrame for the target motif. 
# We check both motif_id and motif_alt_id columns.
matches_df = fimo_df[
    fimo_df['#pattern name'].str.contains(target_motif, na=False)
].copy()

# Select the most relevant columns for sequence inspection
columns_to_keep = ['sequence name', 'start', 'stop', 'strand', 'p-value', 'matched sequence']
filtered_matches = matches_df[columns_to_keep]

# Display the top hits sorted by p-value
filtered_matches = filtered_matches.sort_values(by='p-value')

# Export the filtered matches to a new CSV for permanent record
filtered_matches.to_csv("Data/Motifs_Matches/mmu-miR-466n-5p_matches.csv", index=False)
```
